```python
L=np.argmin(np.array([np.linalg.norm(df_1-centroids[i],axis=1) for i in range(3)]).T,axis=1)

L
```

```python
updated_centroids = np.array([df_1[L == i].mean(axis=0) for i in range(3)])
updated_centroids
```

```python
def update_labels(data, centroids):
    distances = np.array([np.linalg.norm(data - centroid, axis=1) for centroid in centroids]).T
    labels = np.argmin(distances, axis=1)
    return labels
```

```python
def k_means_clustering(data, k, iterations=100):
  df_copy=data.copy()
  centroids=np.array(df_copy.sample(n=k,random_state=42))
  for _ in range(iterations):
    L=update_labels(df_copy,centroids=centroids)
    centroids=np.array([np.mean(df_copy[L==i],axis=0) for i in range(k)])
  return centroids,L

L,centroids=k_means_clustering(df_1,k=3,iterations=2)
print(L)
print(centroids)
```
