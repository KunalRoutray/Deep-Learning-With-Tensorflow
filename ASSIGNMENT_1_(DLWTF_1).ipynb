1. Write a Python code to build a feed-forward neural network for AND, OR, and NAND with a single layer perceptron from
scratch.


```python
import numpy as np
def step_function(z):#step activation function
  return 1 if z>=0 else 0

#perceptron training function
def perceptron_train(X,y,learning_rate=0.1,epochs=5):
   weights=np.zeros(X.shape[1])#initialize weight to zero
   bias=0#initialize bias to zero
   for j in range(epochs):
     print("epochs:",j)
     for i in range(len(X)):
       z=np.dot(weights,X[i])+bias
       print("Z=",z)
       y_pred=step_function(z)#compute output
       print("predicted",y_pred,end='')
       error=y[i]-y_pred

       #update weights an bias
       weights+=learning_rate*error*X[i]
       bias+=learning_rate*error
       #print updates
       print(f"weights:{weights},Bias:{bias}",end='')
       print()
   return weights,bias

#AND gate training data
X=np.array([[0,0],[0,1],[1,0],[1,1]]) #input
y=np.array([0,0,0,1])# and output labels
#train perceptron
weights,bias=perceptron_train(X,y)
def perceptron_predict(X,weights,bias):
  return[step_function(np.dot(weights,x)+bias) for x in X]
outputs=perceptron_predict(X,weights,bias)

#OR gate training data
X=np.array([[0,0],[0,1],[1,0],[1,1]]) #input
y=np.array([0,1,1,1])# and output labels
#train perceptron
weights,bias=perceptron_train(X,y)
def perceptron_predict(X,weights,bias):
  return[step_function(np.dot(weights,x)+bias) for x in X]
outputs=perceptron_predict(X,weights,bias)
print("final prediction :",outputs)

#NAND gate training data
X=np.array([[0,0],[0,1],[1,0],[1,1]]) #input
y=np.array([1,1,1,0])# and output labels
#train perceptron
weights,bias=perceptron_train(X,y)
def perceptron_predict(X,weights,bias):
  return[step_function(np.dot(weights,x)+bias) for x in X]
outputs=perceptron_predict(X,weights,bias)
print("final prediction :",outputs)
```

3. Write a Python code to implement the binary cross-entropy loss function from scratch.


```python
def bce ( actual, predicted,eps=1e-9): #binary cross-entropy
  predicted=np.clip(predicted,eps,1-eps)
  return -(actual*np.log(predicted)+(1-actual)*np.log(1-predicted))
actual=np.array([0,1,1,0])
predicted=np.array([0.1,0.9,0.8,0.7])
print(bce(actual,predicted))
```

4. Write a Python code to implement the Stochastic Gradient Descent Optimization technique.

```python
import numpy as np

def Sigmoid(z):
    return 1/(1+np.exp(-z))

def dsigmoid(a):
    return a*(1-a)

def bce(y,p ,eps=1e-9):
    p=np.clip(p,eps,1-eps)
    return -(y*np.log(p)+(1-y)*np.log(1-p))

X= np.array([[0,0],[0,1],[1,0],[1,1]])
y= np.array([[0],[1],[1],[0]])
rng= np.random.default_rng(0)
W1=np.array([[20.0,20.0],[20.0,20.0]])
b1=np.array([[-10.0,30.0]])
W2=np.array([[20.0],[-20.0]])
b2=np.array([[-10.0]])

#------SGD TRAINING ----
lr=0.1
epochs=10
for ep in range(epochs):
    #shuffle samples each epoch
    idx=rng.permutation(X.shape[0])
    for i in idx:
        xi=X[i:i+1]
        yi=y[i:i+1]
        #forward
        z1=xi @ W1+b1
        a1=Sigmoid(z1)
        z2=a1 @ W2+b2
        a2=Sigmoid(z2)
        #backward(BCE+sigmoid);for single sample N=1
        dz2=(a2-yi)
        dW2=a1.T @ dz2
        db2=dz2
        da1=dz2 @ W2.T
        dz1=da1*dsigmoid(a1)
        dW1=xi.T @ dz1
        db1=dz1
        #SGD update(per sample)
        W1=W1-lr*dW1
        b1=b1-lr*db1
        W2=W2-lr*dW2
        b2=b2-lr*db2
    if ep % 10==0:
        A1 = Sigmoid (X @ W1+b1)
        P  = Sigmoid(A1 @ W2+b2)
        L = bce (y,P).mean()
        print(f"Epoch {ep:4d} | Loss {L:.4f}")
#-----evaluate---
A1 = Sigmoid (X @ W1+b1)
P  = Sigmoid(A1 @ W2+b2)
pred=(P > 0.5).astype(int)
print("\nPredictions (clas | prob):")
for xi,yi,pi in zip(X.astype(int),y.astype(int),P):
    print(f"{tuple(xi)}  -> target ={yi[0]}  pred={int(pi>=0.5)}  p={pi[0]:.3f}")
```

5. Build a multilayer perceptron from scratch using numpy and compare it with Keras.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ----------- KERAS MODEL -----------
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(2, input_dim=2, activation='sigmoid'))  # Hidden layer
model.add(Dense(1, activation='sigmoid'))  # Output layer

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x, y, epochs=10000, verbose=0)

_, keras_accuracy = model.evaluate(x, y, verbose=0)
print(f"\nKeras Model Accuracy: {keras_accuracy * 100:.2f}%")

keras_predictions = model.predict(x)
keras_predictions_rounded = np.round(keras_predictions).astype(int)

print("\nKeras Predictions:")
for i in range(len(x)):
    print(f"Input: {x[i]} -> Predicted: {keras_predictions_rounded[i][0]} , Actual: {y[i][0]}")

# ----------- NUMPY MODEL (SCRATCH) -----------
def sigmoid(X):
    return 1 / (1 + np.exp(-X))

def sigmoid_derivative(X):
    return X * (1 - X)

# Input and Output for XOR
x2 = np.array([[0,0],
               [0,1],
               [1,0],
               [1,1]])

y2 = np.array([[0],
               [1],
               [1],
               [0]])

np.random.seed(42)
input_layer_neurons = x2.shape[1]
hidden_layer_neurons = 2
output_layer_neurons = 1

W1 = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
b1 = np.random.uniform(size=(1, hidden_layer_neurons))
W2 = np.random.uniform(size=(hidden_layer_neurons, output_layer_neurons))
b2 = np.random.uniform(size=(1, output_layer_neurons))

lr = 0.1
epochs = 10000

for i in range(epochs):
    # Forward propagation
    hidden_input = np.dot(x2, W1) + b1
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, W2) + b2
    y_pred2 = sigmoid(final_input)

    # Error
    error = y2 - y_pred2

    # Backpropagation
    d_y_pred = error * sigmoid_derivative(y_pred2)
    error_hidden_layer = d_y_pred.dot(W2.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_output)

    # Update weights
    W2 += hidden_output.T.dot(d_y_pred) * lr
    b2 += np.sum(d_y_pred, axis=0, keepdims=True) * lr
    W1 += x2.T.dot(d_hidden_layer) * lr
    b1 += np.sum(d_hidden_layer, axis=0, keepdims=True) * lr

    if i % 1000 == 0:
        loss = np.mean(np.abs(error))
        print(f"Epoch {i}: Loss={loss}")

# Final forward pass for predictions
hidden_input = np.dot(x2, W1) + b1
hidden_output = sigmoid(hidden_input)
final_input = np.dot(hidden_output, W2) + b2
final_predictions = sigmoid(final_input)
numpy_predictions_rounded = np.round(final_predictions)

print("\nNumPy Model Predictions:")
print(np.round(final_predictions))

# ----------- FINAL COMPARISON -----------
print("\n===== COMPARISON =====")
print(f"Keras Accuracy: {keras_accuracy * 100:.2f}%")
print("Keras Predicted Outputs:", keras_predictions_rounded.flatten())
print("NumPy Predicted Outputs:", numpy_predictions_rounded.flatten())
```

6. Write a Python code to classify the XOR function and use sigmoid, ReLU, and LeakyReLU in the hidden layer of a neural
network. Discuss the accuracy of different activation functions used in the hidden layer. Also, show the comparative analysis in
a graph


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.optimizers import SGD
from sklearn.metrics import accuracy_score

#XOR dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

#NumPy implementation (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize weights
np.random.seed(42)
W1 = np.random.uniform(size=(2,4))
b1 = np.random.uniform(size=(1,4))
W2 = np.random.uniform(size=(4,1))
b2 = np.random.uniform(size=(1,1))

lr = 0.1
epochs = 10000
for _ in range(epochs):
    H_in = X.dot(W1) + b1
    H_out = sigmoid(H_in)
    O_in = H_out.dot(W2) + b2
    O_out = sigmoid(O_in)

    error = y - O_out
    dO = error * sigmoid_derivative(O_out)
    dH = dO.dot(W2.T) * sigmoid_derivative(H_out)

    W2 += H_out.T.dot(dO) * lr
    b2 += np.sum(dO, axis=0, keepdims=True) * lr
    W1 += X.T.dot(dH) * lr
    b1 += np.sum(dH, axis=0, keepdims=True) * lr

# Predictions
H_out = sigmoid(X.dot(W1) + b1)
O_out = sigmoid(H_out.dot(W2) + b2)
numpy_preds = np.round(O_out).astype(int)
numpy_acc = accuracy_score(y, numpy_preds)
print("NumPy Sigmoid Predictions:", numpy_preds.ravel())
print(f"NumPy Sigmoid Accuracy: {numpy_acc*100:.2f}%")

#Keras implementation (Sigmoid, ReLU, LeakyReLU)
def train_keras_xor(activation='sigmoid'):
    model = Sequential()
    if activation == 'leakyrelu':
        model.add(Dense(4, input_dim=2))
        model.add(LeakyReLU(alpha=0.1))
    else:
        model.add(Dense(4, input_dim=2, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=SGD(learning_rate=0.1),
                  loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(X, y, epochs=1000, verbose=0)

    y_pred = np.round(model.predict(X))
    acc = accuracy_score(y, y_pred)
    return y_pred, acc

# Train models with different activations
activations = ['sigmoid', 'relu', 'leakyrelu']
results = {}
for act in activations:
    pred, acc = train_keras_xor(act)
    results[act] = acc
    print(f"Activation: {act}, Predictions: {pred.ravel()}, Accuracy: {acc*100:.2f}%")

#Comparative Accuracy Graph
plt.bar(results.keys(), results.values(), color=['skyblue', 'salmon', 'lightgreen'])
plt.ylabel('Accuracy')
plt.title('XOR Classification Accuracy with Different Hidden Layer Activations')
plt.ylim(0,1.1)
plt.show()
```

7. Write code to define a simple neural network in TensorFlow. Use the MNIST dataset, reshape the data to 610,000 x 784,
normalize the input, and build and train the model using the SGD optimizer and categorical cross-entropy as the loss function to
compute the accuracy. Also, find the test accuracy.


```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
EPOCHS=200
BATCH_SIZE=128
VERBOSE=1
NB_CLASSES=10
N_HIDDEN=128
VALIDATION_SPLIT=0.2

mnist=keras.datasets.mnist
(X_train, Y_train), (X_test,Y_test)=mnist.load_data()
RESHAPED=784
X_train= X_train.reshape(60000,RESHAPED)
X_test=X_test.reshape(10000,RESHAPED)
X_train=X_train.astype('float32')/255
X_test=X_test.astype('float32')/255
Y_train=tf.keras.utils.to_categorical(Y_train,NB_CLASSES)
Y_test=tf.keras.utils.to_categorical(Y_test,NB_CLASSES)

model=keras.Sequential([
    keras.layers.Input(shape=(RESHAPED,)),
    keras.layers.Dense( N_HIDDEN,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(N_HIDDEN,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(NB_CLASSES,activation='softmax')
])

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
es=keras.callbacks.EarlyStopping(monitor="val_accuracy",patience=5)
lr=keras.callbacks.ReduceLROnPlateau(monitor="val_loss",factor=0.5)
history=model.fit(
    X_train,Y_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_split=VALIDATION_SPLIT,
    verbose=VERBOSE,
    callbacks=[es,lr]
)
test_loss,test_acc=model.evaluate(X_test,Y_test,verbose=0)
print(f"Test accuracy: {test_acc:.4f} | Test loss:{test_loss:.4f}")
```

8. Train the MLP as specified in Q7 with regularization (dropout and L2), different optimizers, and discuss a detailed comparative
analysis.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import SGD, Adam, RMSprop
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 28*28).astype('float32') / 255.0
X_test = X_test.reshape(X_test.shape[0], 28*28).astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Function to build model
def build_model(optimizer_name='sgd', l2_lambda=0.0, dropout_rate=0.0):
    if optimizer_name.lower() == 'sgd':
        optimizer = SGD(learning_rate=0.1)
    elif optimizer_name.lower() == 'adam':
        optimizer = Adam(learning_rate=0.001)
    elif optimizer_name.lower() == 'rmsprop':
        optimizer = RMSprop(learning_rate=0.001)
    else:
        optimizer = SGD(learning_rate=0.1)

    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,), kernel_regularizer=l2(l2_lambda)))
    if dropout_rate > 0:
        model.add(Dropout(dropout_rate))
    model.add(Dense(10, activation='softmax'))

    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Different optimizers and regularization settings
results = {}
configs = [
    {'name': 'SGD', 'l2': 0.0, 'dropout': 0.0},
    {'name': 'SGD+L2', 'l2': 0.01, 'dropout': 0.0},
    {'name': 'SGD+Dropout', 'l2': 0.0, 'dropout': 0.2},
    {'name': 'Adam', 'l2': 0.0, 'dropout': 0.0},
    {'name': 'Adam+Dropout+L2', 'l2': 0.01, 'dropout': 0.2},
    {'name': 'RMSprop', 'l2': 0.0, 'dropout': 0.0}
]

for cfg in configs:
    print(f"Training model: {cfg['name']} | L2={cfg['l2']} | Dropout={cfg['dropout']}")
    model = build_model(
        optimizer_name=cfg['name'].split('+')[0],
        l2_lambda=cfg['l2'],
        dropout_rate=cfg['dropout']
    )
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0, validation_split=0.1)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    results[cfg['name']] = test_acc
    print(f"Test Accuracy: {test_acc * 100:.2f}%\n")

# Comparative Accuracy Graph
plt.figure(figsize=(10,5))
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Test Accuracy')
plt.ylim(0,1)
plt.title('MNIST MLP: Accuracy Comparison with Regularization and Optimizers')
plt.xticks(rotation=20)
plt.show()
```
